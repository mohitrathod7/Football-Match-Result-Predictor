# Cleaning Data

````python
matches.dtypes
````

ML algorithms cannot work with "object" data so we need to convert object to numeric predictors.

````python
# Convert existing date column to datetime column
matches["date"] = pd.to_datetime(matches["date"])
````

# Creating Predictor variables

````python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
````

````python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
````

````python
# time is 16:30, we will remove :30 and keep only 16, and convert to integer
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
````

````python
# Monday: 0 ... Sunday: 6
matches["day_code"] = matches["date"].dt.dayofweek
````

````python
# Win: 1 and Loss,Draw: 0
matches["target"] = (matches["result"] == "W").astype("int")
````

# Modeling

> This algorithm can pickup non-linearities in data. It is a series of decision trees, higher number of tress (n_estimators) more accuracy and more time to run.

> For example, Opponent code 12 or 16 does not mean that opponent 16 is more difficult opponent 12, it is just a code for the opponents.

````python
from sklearn.ensemble import RandomForestClassifier
````

````python
# random_state = 1 means we will get same results any time we run this as long as the data is same
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
````

````python
train = matches[matches["date"] < '2022-01-01']
````

````python
test = matches[matches["date"] > '2022-01-01']
````

````python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
````

````python
rf.fit(train[predictors], train["target"])
````

````python
# Generate predictions from test data
preds = rf.predict(test[predictors])
````

Now we will determine accuracy of the model

# Improving accuracy of the model

To improve accuracy we will look into more recent averages of the team such as:

Compute rolling averages, how the team did in past few games, how many goals were scored against them, etc.

Suppose if we are in week 4 for Manchester city, so we need to know how did the team performed in 3 weeks before this week (like won all last 3 matches), and get average of it.

````python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group

# from views.functions import rolling_averages
````

gf: goals for

ga: goals against

sh: shots taken

sot: shots on target

dist: total distance each shot traveled

fk: free kicks

pk: penalty kicks

pkatt: penalty kicks attempts

````python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
````

````python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling
````

````python
matches_rolling = matches_rolling.droplevel("team") # drop extra index level "team" by position 0
matches_rolling
````

````python
# Reset index from 0 to end
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
````

# Retraining our Model

````python
def make_predictions(data, predictors):
  train = data[data["date"] < '2022-01-01']
  test = data[data["date"] > '2022-01-01']
  rf.fit(train[predictors], train["target"])
  preds = rf.predict(test[predictors])
  combined = pd.DataFrame(dict(actual=test["target"], prediction=preds), index=test.index)
  precision = precision_score(test["target"], preds)
  return combined, precision
````

````python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
````

# Predict new match

Predicts the outcome of a new match based on the provided information.

**Args:**

    team: The name of the team.
    opponent: The name of the opponent.
    venue: The venue of the match ("Home" or "Away").
    date: The date of the match as a string in YYYY-MM-DD format.
    hour: The hour of the match.
    rf_model: The trained RandomForestClassifier model.
    predictors: The list of predictor variables.

**Returns:**

    A dictionary containing the predicted outcome (1 for win, 0 for loss/draw)
    and the probabilities for win and loss/draw.


````python
def predict_new_match_outcome(team, opponent, venue, date, hour, rf_model, predictors):
  team_data = matches_rolling[matches_rolling["team"] == team]
  team_data = team_data.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
  team_data = team_data.droplevel("team")
  team_data.index = range(team_data.shape[0])

  gf_rolling = team_data["gf_rolling"].iloc[-1]
  ga_rolling = team_data["ga_rolling"].iloc[-1]
  sh_rolling = team_data["sh_rolling"].iloc[-1]
  sot_rolling = team_data["sot_rolling"].iloc[-1]
  dist_rolling = team_data['dist_rolling'].iloc[-1]
  fk_rolling = team_data["fk_rolling"].iloc[-1]
  pk_rolling = team_data["pk_rolling"].iloc[-1]
  pkatt_rolling = team_data["pkatt_rolling"].iloc[-1]

  new_match = pd.DataFrame({
      'team': [team],
      'opponent': [opponent],
      'venue': [venue],
      'date': [pd.to_datetime(date)],
      'time': ['{hour}:00'.format(hour=hour)],
      'gf': [np.nan],
      'ga': [np.nan],
      'sh': [np.nan],
      'sot': [np.nan],
      'dist': [np.nan],
      'fk': [np.nan],
      'pk': [np.nan],
      'pkatt': [np.nan],
      'gf_rolling': [gf_rolling],
      'ga_rolling': [ga_rolling],
      'sh_rolling': [sh_rolling],
      'sot_rolling': [sot_rolling],
      'dist_rolling': [dist_rolling],
      'fk_rolling': [fk_rolling],
      'pk_rolling': [pk_rolling],
      'pkatt_rolling': [pkatt_rolling]
  })

  new_match["venue_code"] = new_match["venue"].astype("category").cat.codes
  new_match["opp_code"] = new_match["opponent"].astype("category").cat.codes
  new_match["hour"] = new_match["time"].str.replace(":.+", "", regex=True).astype("int")
  new_match["day_code"] = new_match["date"].dt.dayofweek

  prediction = rf_model.predict(new_match[predictors])
  probabilities = rf_model.predict_proba(new_match[predictors])

  return {
      'prediction': prediction[0],
      'win_probability': probabilities[0][1],
      'loss_draw_probability': probabilities[0][0],
  }
````

````python
# Usage:
team_name = "Liverpool"
opponent_name = "Manchester"
venue_name = "Home"
match_date = "2023-11-18"
match_hour = 16

prediction_result = predict_new_match_outcome(
    team_name,
    opponent_name,
    venue_name,
    match_date,
    match_hour,
    rf,
    predictors + new_cols
)

print(prediction_result)
````
